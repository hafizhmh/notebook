```
import pandas as pd
import io
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```
# Penyiapan data
names=['x_coor', 'y_coor']
df1 =  pd.read_csv('https://raw.githubusercontent.com/hafizhmh/etc/main/number_1_point_1.csv', names=names)
df2 =  pd.read_csv('https://raw.githubusercontent.com/hafizhmh/etc/main/number_1_point_2.csv', names=names)
```

```
# Visualisasi data
ax1 = df1.plot(kind='scatter', x='x_coor', y='y_coor')
ax1.title.set_text('Point 1')

ax2 = df2.plot(kind='scatter', x='x_coor', y='y_coor')
ax2.title.set_text('Point 2')
```

```
# Kalkulasi sentroid
# Dari gambar, nampaknya ada 3 buah kluster
kmeans1 = KMeans(n_clusters=3)
kmeans1.fit(df1)
print("Set 1 centroid:")
print(kmeans1.cluster_centers_)
print(kmeans1.labels_)
#print()

kmeans2 = KMeans(n_clusters=3)
kmeans2.fit(df2)
print("Set 2 centroid:")
print(kmeans2.cluster_centers_)
print(kmeans2.labels_)
#print()
```

```
# Visualisasi kluster dan sentroid
ax1 = df1.plot(kind='scatter', x='x_coor', y='y_coor', c=kmeans1.labels_, cmap='rainbow')
ax1.scatter(kmeans1.cluster_centers_[:,0] ,kmeans1.cluster_centers_[:,1], color='black')
ax1.title.set_text('Point 1')

ax2 = df2.plot(kind='scatter', x='x_coor', y='y_coor', c=kmeans2.labels_, cmap='rainbow')
ax2.scatter(kmeans1.cluster_centers_[:,0] ,kmeans1.cluster_centers_[:,1], color='black')
ax2.title.set_text('Point 2')
```

```
# Prediksi: point 1
to_be_predicted = np.array([[0,0], [0,1], [1,0], [1,1], [-2,1]])
k1_p = kmeans1.predict(to_be_predicted)
print(k1_p)

ax1 = df1.plot(kind='scatter', x='x_coor', y='y_coor', c=kmeans1.labels_, cmap='rainbow')
ax1.scatter(kmeans1.cluster_centers_[:,0] ,kmeans1.cluster_centers_[:,1], color='black')
ax1.scatter(to_be_predicted[:,0], to_be_predicted[:,1], color='orange')

for i, txt in enumerate(k1_p):
    ax1.annotate("{}: cls {}".format(i, txt), (to_be_predicted[i,0], to_be_predicted[i,1]))
    print("{}: ({},{}) -> Cluster {}".format(i, to_be_predicted[i,0], to_be_predicted[i,1], txt))
ax1.title.set_text('Point 1')
```

```
# Prediksi: point 2
to_be_predicted = np.array([[0,0], [0,1], [1,0], [1,1], [-2,1]])
k2_p = kmeans2.predict(to_be_predicted)
print(k2_p)

ax2 = df2.plot(kind='scatter', x='x_coor', y='y_coor', c=kmeans2.labels_, cmap='rainbow')
ax2.scatter(kmeans1.cluster_centers_[:,0] ,kmeans1.cluster_centers_[:,1], color='black')
ax2.scatter(to_be_predicted[:,0], to_be_predicted[:,1], color='orange')

for i, txt in enumerate(k2_p):
    ax2.annotate("{}: cls {}".format(i, txt), (to_be_predicted[i,0], to_be_predicted[i,1]))
    print("{}: ({},{}) -> Cluster {}".format(i, to_be_predicted[i,0], to_be_predicted[i,1], txt))
ax2.title.set_text('Point 2')
```
