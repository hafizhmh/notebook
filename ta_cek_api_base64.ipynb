```
import base64
import requests
from google.colab import files
from IPython.display import Image
from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image
```

```
# Query
NRKB = "F 2825 BH"
URL = "https://stnk-api-ta.tech/api/perpanjangan?nrkb="+NRKB
r = requests.get(url = URL)
data = r.json()
```

```
# Decoding base64
gambar = data[0]['fotoKTP']
imgdata = base64.b64decode(gambar)
filename = 'a.jpg'  # I assume you have a way of picking unique filenames
with open(filename, 'wb') as f:
    f.write(imgdata)
# files.download('a.jpg')
pil_im = Image.open('a.jpg', 'r')
imshow(np.asarray(pil_im))
```
